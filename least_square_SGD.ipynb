## Least square stochastic gradient descent

```python
w_initial = np.random.rand(num_features)
batch_size = 1
max_iters = 100
gamma = 0.01

weights, loss = least_squares_SGD (y, tx, w_initial, batch_size, max_iters, gamma)
```

## Look for the best value of gamma

```python
w_initial = np.random.rand(num_features)
batch_size = 1
max_iters = 100
gammas = np.arange(0.0005, 0.005, 0.0001)
ws = []
losses = []

for ind, gamma in enumerate(gammas):
    w, l = least_squares_SGD(y, tx, w_initial, batch_size, max_iters, gamma, printing=False)
    ws.append(w)
    losses.append(l)

loss = np.amin(losses)
weights = ws[np.argmin(losses)]
gamma_star = gammas[np.argmin(losses)]

plt.plot(gammas, losses, color='k')
plt.plot(gamma_star, loss, 'r*', markersize=15, label="Minimal loss")
plt.xlabel('Gamma')
plt.ylabel('Loss')
plt.title("Loss in function of gamma")
plt.legend()
plt.show()

print(" Best value of gamma = {g} \n Loss = {l} \n Weights = {we}".format(
    g=gamma_star, l=loss, we = weights))
```
